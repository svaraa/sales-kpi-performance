```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/Superstore.csv", encoding='latin1')
df.head()
```

```python
df.info()
df.describe(include='all')
df.isna().sum()
```

```python
# basic cleaning
df.columns = (
    df.columns
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("-", "_")
    .str.lower()
)
df.head()
```

```python
# converting dates
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
# dropping duplicates
df.drop_duplicates(inplace=True)
```

```python
# handling missing values
if 'postal_code' in df.columns:
    df['postal_code'] = df['postal_code'].fillna(0)

# dropping rows missing sales/profit/order date
df = df.dropna(subset=['sales', 'profit', 'order_date'])
```

```python
# feature engineering
# year
df['year'] = df['order_date'].dt.year

# month number
df['month'] = df['order_date'].dt.month

# month name
df['month_name'] = df['order_date'].dt.strftime("%B")

# profit margin
df['profit_margin'] = df['profit'] / df['sales']

# avg order value
df['aov'] = df['sales'] / df.groupby('order_id')['order_id'].transform('count')
```

```python
text_cols = ['category', 'sub_category', 'segment', 'country', 'city', 'region']

for col in text_cols:
    if col in df.columns:
        df[col] = df[col].str.strip()
```

```python
df.info()
df.head()
```

```python
df.to_csv("../data/superstore_clean.csv", index=False)
```
